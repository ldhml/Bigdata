```python
#승차인원이 최대인 역을 찾고 그 값을 찾아라
import csv
file=open('/content/subwayfree.csv','r',encoding='cp949')
data=csv.reader(file)
next(data)

mx=-1 #mx는 max의 변수 처음부터 100으로 주면 밑에 결과가 다 작을것이기 때문에 적은 -1로 함. 예를들어 제일 작은값이면 -1이면 안됨. 
      #1000이던 내가 생각한 큰 값을 적는다.
rate=0 #rate는 변수이고  저장하는 변수명이다. 예를들어 4인덱스 + 6인덱스 분에 4인덱스를 저장하는 공간이 rate라는 것임.
mx_station=''

for row in data:
  for i in range(4,8): # 엑셀에는 써있는것이 숫자가 아니니까 첫번째부터 0번 1번 2번 이런식으로 생각해야함.
    row[i]=int(row[i])
  if(row[6]==0 and row[4]+row[6]<100000): #만약에 6번째 인덱스가 6이 0이면 그냥 넘어가라 또 4번, 6번 인덱스가 100000보다 작으면 다음으로 넘어가라.
   continue
  rate=row[4]/(row[4]+row[6]) #예를들어 첫 번째 인덱스를 계산했는데 값이 80이 나왔다면
  if rate > mx: #바로 윗줄에서 구한 80을 rate에 대입해서 저장되어있는 최대값이랑 비교하는것.
    mx=rate #계산한 값이 최대값이면 mx에 저장되는것.
    mx_station=row[3]=''+row[1]
print(mx_station,round(mx*100,2))
```

```python
#유무임 승하차 인원이 가장 많은 역은 어디일까?
import csv
file=open('/content/subwayfree.csv','r',encoding='cp949')
data=csv.reader(file)
next(data)

mx=[-1]*4 # [-1 -1 -1 -1] 이렇게 저장되어있음 각각 mx[0],mx[1],mx[2],mx[3]이렇게 저장되어있는것
mx_station=['']*4
label=['유임승차','유임하차','무임승차','무임하차']

for row in data:
  for i in range(4,8): #인덱스 4번부터 8번까지
    row[i]=int(row[i]) #인티저로 바꾸고싶음.
    if(row[i]>mx[i-4]):#row[4]>mx[0] 4번째 인덱스를 첫번째꺼랑 비교하기 위해 -4해준것임.
      mx[i-4]=row[i]
      mx_station[i-4]=row[3]+' '+row[1]

for i in range(4):
  print(label[i]+": "+mx_station[i],mx[i])
```

```python
#모든 지하철역에서 시간대별 승하차 인원을 모두 더하면
import csv
import matplotlib.pyplot as plt
file=open('/content/subwaytime.csv','r',encoding='cp949')
data=csv.reader(file)
next(data)
next(data) #next는 시작위치를 아랫줄로 바꾸는 것. 왜 2번썼냐면 엑셀파일에 1,2줄이 글자와 공백이라

s_in=[0] * 24 #모든 지하철의 승하차를 구하는 것이니 24임.
s_out=[0] * 24

#밑에 for문 얘기임. for문 안에 for문이니까 첫번째 for문안에서 첫 번째 역을 선정함. 그 밑에 for문에서 승하차 나옴. 
#위에 for문 다돌고 끝나면 다시 위로 올라가서 첫 번째 for문으로 돌아감. 그럼 다음역 두 번째 역을 넣고 그다음 for으로 들어가서 다시 승하차를 계산함.
for row in data:
  row[4:]=map(int,row[4:]) #4번 인덱스부터 끝까지 /map함수는 괄호의 두 번째 인덱스를 int로 바꾼다는 함수
  for i in range(24): #0부터 23까지(23까지면 24로 표기)
    s_in[i]+=row[4+i*2] #첫 번째로는 0번째를 i에 집어 넣음. 계산해보면 4임. 그러면 4번째 인덱스가 s_in 즉 4번째 인덱스에 있는 4시 승차인원이 됨.
    s_out[i]+=row[5+i*2] #첫 번째끝나면 다음 인덱스에 1 들어가서 for문돌음. 그러면 s-in에는 4,6,8시 이런식으로 승차인원이 들어감. 마지막 새벽3시까지 for문 돔.

for j in range(24): #위에 문단for문에 승하차 인원이 저장되어있으니 잘 저장되어있는지 확인을 위해 넣어놓은거임. 
  print(s_in[j],s_out[j]) #숫자만 출력할거면 여기서 끝내고 그래프로 나타내면 밑에 추가

#그래프로 나타내기 위에는 숫자만 출력하는거임.
plt.title('# of passengers getting IN and OUT by time')
plt.plot(s_in,label="IN")
plt.plot(s_out,label="out")
plt.xticks(range(24),range(4,28)) #4,28은 4번째 인덱스부터 새벽3시인 27까지 표기로는 28로 표기함. 
plt.legend()
plt.show()
```

```python
#우리나라 몇위?
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path='/content/drinks.csv'
drinks=pd.read_csv(file_path)

drinks['continent']=drinks['continent'].fillna('OT')

#total_servings 피처를 생성, 마지막 열에 생성됨 #엑셀 파일 마지막 열에 total_servings라는 칸이 따로 생김.
drinks['total_servings']=drinks['beer_servings']+drinks['wine_servings']+drinks['spirit_servings']

#술 소비량 대비 알콜 비율 피처를 생성 #엑셀 파일 마지막 열에 alcohol_rate이라는칸이 따로 생김.
drinks['alcohol_rate']=drinks['total_litres_of_pure_alcohol']/drinks['total_servings']

#결측값을 0으로 #알코올을 아예안먹는 나라가 있으면 분모가 0이 되므로 결측값이 있으면 0으로 채워라.
drinks['alcohol_rate']=drinks['alcohol_rate'].fillna(0)

#순위 정보를 생성
country_with_rank=drinks[['country','alcohol_rate']]
country_with_rank=country_with_rank.sort_values(by=['alcohol_rate'],ascending=0)
country_with_rank.head(5)
```

```python
#서울에서 경기도로 이동한 인구
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.simplefilter("ignore")

file_path='/content/inoutpeople.xlsx'
df=pd.read_excel(file_path,engine='openpyxl',header=0)

df=df.fillna(method='ffill')
#서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask=(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')

#true인 부분만 df_seoul에 저장, 즉 서울에서 다른 지역으로 전출 하는 데이터만
df_seoul=df[mask] #true인 애들만 df_seoul에 저장된다.

#필요없으니 버리고
df_seoul=df_seoul.drop(['전출지별'],axis=1)

#inplace=True,df_seoul을 바꿔라 false이면 연결된 df_seoul가 바뀌지 않고 객체에 할당해줘야함
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)

#'전입지'가 df_seoul의 행인덱스로 지정
df_seoul.set_index('전입지',inplace=True)

#서울에서 경기도로 이동한 인구 데이터 값만 선택, loc -->열선택
sr_one=df_seoul.loc['경기도']
print(sr_one)

#x,y축 데이터를 plot함수에 입력
plt.plot(sr_one.index,sr_one.values)
plt.rc('font',family='NanumBarunGothic')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.simplefilter("ignore")

file_path='/content/inoutpeople.xlsx'
df=pd.read_excel(file_path,engine='openpyxl',header=0)

df=df.fillna(method='ffill')

#서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask=(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul=df[mask]
df_seoul=df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)

#서울에서 충청남도, 경상북도, 강원도 전라남도 로 이동한 인구 데이터 값만 선택
df_4=df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4=df_4.transpose()

#스타일 서식 지정
plt.style.use('ggplot')

df_4.index=df_4.index.map(int)

#면적 그래프 그리기
df_4.plot(kind='area',stacked=True, alpha=0.2,figsize=(20,10))

plt.title('서울 -> 타시도 인구 이동',size=30,color='brown',weight='bold')
plt.ylabel('이동 인구 수',size=20,color='blue')
plt.xlabel('기간',size=20,color='blue')
plt.legend(loc='best',fontsize=15)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.simplefilter("ignore")

file_path='/content/inoutpeople.xlsx'
df=pd.read_excel(file_path,engine='openpyxl',header=0)

df=df.fillna(method='ffill')

#서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask=(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul=df[mask]
df_seoul=df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)

#서울에서 충청남도, 경상북도, 강원도, 전라남도로 이동한 인구 데이터 값만 선택
#col_year에 해당하는 부분만 df_4에 저장후 transpose
col_years=list(map(str,range(2010,2018)))
df_4=df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]
df_4=df_4.transpose()

plt.style.use('ggplot')

#데이터프레임의 인덱스를 정수형으로 변경
df_4.index=map(int,df_4.index)

#막대 그래프 그리기
df_4.plot(kind='bar',figsize=(20,10),width=0.7,color=['orange','green','skyblue','blue'])

plt.title('서울 -> 타시도 인구 이동',size=30)
plt.ylabel('이동 인구 수',size=30)
plt.xlabel('기간',size=20)
plt.ylim(5000,30000)
plt.legend(loc='best',fontsize=15)

plt.show()
```
