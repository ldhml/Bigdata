```python
import csv
import matplotlib.pyplot as plt
 
f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
 
aug = []                          
jan = []                          

for row in data :
    month = row[0].split('-')[1]  
    if row[-1] != '' :
        if month == '08':               
            aug.append(float(row[-1]))  
        if month == '01':               
            jan.append(float(row[-1]))  

plt.hist(aug, bins=100, color ='r', label='Aug')
plt.hist(jan, bins=100, color ='b', label='Jan')
plt.legend()
plt.show()
```

```python
#우리동네인구 남/녀 비중을 보기 위해 누적연산 수행
import csv
import matplotlib.pyplot as plt
f=open('/content/gender.csv','r',encoding='cp949')
data=csv.reader(f)
size=[]
m = f = 0

for row in data:
  if'신도림'in row[0]:
    for i in range(0,101):
      m+=int(row[i+3])
      f+=int(row[-(i+1)])

size.append(m)
size.append(f)

color=['magenta','cyan']
plt.title('Gender Distribution in Sindorim')
plt.axis('equal')
plt.pie(size,labels=['M','F'],autopct='%.1f%%',colors=color)
plt.show()
```

```python
#연령대별 성별 비율 산점도로 표현하기(원의 크기는 인구수의 제곱근)하기
import csv
import math
import matplotlib.pyplot as plt

f=open('/content/gender.csv','r',encoding='cp949')
data=csv.reader(f)
m=[]
f=[]
size=[]
name=input('원하는 동 입력: ')

for row in data:
  if name in row[0]:
    for i in range(3,104):
      temp1=int(row[i].replace(",",""))
      temp2=int(row[i+103].replace(",",""))
      m.append(temp1)
      f.append(temp2)
      size.append(math.sqrt(temp1+temp2))
    break
plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.scatter(m,f,s=size,c=range(101),alpha=0.5,cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g')
plt.xlabel('Male population')
plt.ylabel('Female population')
plt.show()
```

```python
import csv
file=open('/content/subwayfree.csv','r',encoding='cp949')
data=csv.reader(file)
next(data)

mx=-1
rate=0
mx_station=''

for row in data:
  for i in range(4,8):
    row[i]=int(row[i])
  if(row[6]==0 and row[4]+row[6]<100000):
   continue
  rate=row[4]/(row[4]+row[6])
  if rate > mx:
    mx=rate
    mx_station=row[3]=''+row[1]
print(mx_station,round(mx*100,2))
```

```python
#유무임 승하차 인원이 가장 많은 역은 어디일까?
import csv
file=open('/content/subwayfree.csv','r',encoding='cp949')
data=csv.reader(file)
next(data)

mx=[-1]*4
mx_station=['']*4
label=['유임승차','유임하차','무임승차','무임하차']

for row in data:
  for i in range(4,8):
    row[i]=int(row[i])
    if(row[i]>mx[i-4]):
      mx[i-4]=row[i]
      mx_station[i-4]=row[3]+' '+row[1]

for i in range(4):
  print(label[i]+": "+mx_station[i],mx[i])
```
